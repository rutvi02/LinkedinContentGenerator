# Function to extract all the fields from json file

```python

def flatten_linkedin_post(post):
    return {
        "urn": post.get("urn"),
        "text": post.get("text"),
        "url": post.get("url"),
        "date": post.get("posted_at", {}).get("date"),
        "relative_time": post.get("posted_at", {}).get("relative"),
        "timestamp": post.get("posted_at", {}).get("timestamp"),
        "author_first_name": post.get("author", {}).get("first_name"),
        "author_last_name": post.get("author", {}).get("last_name"),
        "author_headline": post.get("author", {}).get("headline"),
        "author_username": post.get("author", {}).get("username"),
        "author_profile_url": post.get("author", {}).get("profile_url"),
        "author_profile_picture": post.get("author", {}).get("profile_picture"),
        "likes": post.get("stats", {}).get("like"),
        "support": post.get("stats", {}).get("support"),
        "love": post.get("stats", {}).get("love"),
        "insight": post.get("stats", {}).get("insight"),
        "celebrate": post.get("stats", {}).get("celebrate"),
        "comments": post.get("stats", {}).get("comments"),
        "reposts": post.get("stats", {}).get("reposts"),
        "total_reactions": post.get("stats", {}).get("total_reactions"),
        "media_type": post.get("media", {}).get("type"),
        "media_url": post.get("media", {}).get("url"),
        "media_image_url": post.get("media", {}).get("images", [{}])[0].get("url") if post.get("media") else None,
        "media_image_width": post.get("media", {}).get("images", [{}])[0].get("width") if post.get("media") else None,
        "media_image_height": post.get("media", {}).get("images", [{}])[0].get("height") if post.get("media") else None,
        "pagination_token": post.get("pagination_token")
    }

```
